```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score


from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score
from sklearn.preprocessing import LabelBinarizer

# Supongamos que tu DataFrame final se llama 'df'
# Las columnas son OTUs y variables de metadatos, con una columna 'target' que es lo que quieres predecir

# Cargar tus datos (este paso dependerá de cómo tengas almacenados tus datos)
# df = pd.read_csv('tu_archivo.csv')

# Separar las variables independientes (X) y la variable dependiente (y)
X = df_combinado.drop('inflamacion', axis=1)  # Remplaza 'target' con el nombre de tu columna objetivo
y = df_combinado['inflamacion']

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Crear y entrenar el modelo de árbol de decisión
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)

# Predecir el conjunto de prueba
y_pred = clf.predict(X_test)

# Evaluar el modelo
accuracy = accuracy_score(y_test, y_pred)
print(f'Precisión del modelo: {accuracy}')
```

```python
# Calcular métricas de rendimiento
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
# Para el AUC, necesitas etiquetas binarias; si es un problema multiclase, ajusta esto apropiadamente.
if len(y.unique()) == 2:  # Problema binario
    lb = LabelBinarizer()
    y_test_bin = lb.fit_transform(y_test)
    y_pred_bin = lb.transform(y_pred)
    print("AUC ROC:", roc_auc_score(y_test_bin, y_pred_bin))

# Obtener e imprimir la importancia de las características
importancias = clf.feature_importances_
caracteristicas = X.columns
print("Importancia de las Características:")
for caracteristica, importancia in zip(caracteristicas, importancias):
    print(f"{caracteristica}: {importancia}")
```

```python
from statsmodels.stats.proportion import proportions_ztest

def z_test_two_proportions(p1, n1, p2, n2):
    # Perform Z-test for two proportions
    z_statistic, p_value = proportions_ztest([p1 * n1, p2 * n2], [n1, n2])
    
    return z_statistic, p_value

# Example data
TreatmentA = otu1_si  # Percentages for Treatment A
TreatmentB = otu1_no     # Percentages for Treatment B

# Calculate sample sizes
n1 = len(TreatmentA)  # Number of observations in Treatment A
n2 = len(TreatmentB)  # Number of observations in Treatment B

# Calculate sample proportions (convert percentages to proportions)
p1 = sum(TreatmentA) / n1  # Proportion for Treatment A
p2 = sum(TreatmentB) / n2  # Proportion for Treatment B

# Perform Z-test for two proportions
z_statistic, p_value = z_test_two_proportions(p1, n1, p2, n2)

# Output results
print("Z-statistic:", z_statistic)
print("p-value:", p_value)

# Interpret results
alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis: There is evidence of a difference in the proportions between Treatment A and Treatment B.")
else:
    print("Fail to reject null hypothesis: There is no evidence of a difference in the proportions between Treatment A and Treatment B.")
```
